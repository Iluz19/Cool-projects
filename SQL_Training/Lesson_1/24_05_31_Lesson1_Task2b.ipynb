```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
user_details = pd.read_csv ("/Users/eilonzohar/dumps/user_details.csv")
user_details.head (10)
```

```python
first_names_sorted = np.sort (list (user_details['first_name'].unique())) [::-1]
fig, ax = plt.subplots (figsize= (20, 7));
plt.grid();
plt.hist (first_names_sorted, orientation="horizontal")
plt.xlabel ("Number of users")
plt.ylabel ("Name")
plt.show ()
```
